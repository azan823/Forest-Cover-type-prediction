```python
x = df.drop('Cover_Type',axis=1)
y = df['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_train,y_train)
y_pred = lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,recall_score
print(f"accuarcy{accuracy_score(y_test,y_pred)}")
print(f"f1 score{f1_score(y_test,y_pred,average='weighted')}")
print(f"recall score{recall_score(y_test,y_pred,average='weighted')}")

```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred1 = dt.predict(x_test)
```

```python
print(f"accuarcy{accuracy_score(y_test,y_pred1)}")
print(f"f1 score{f1_score(y_test,y_pred1,average='weighted')}")
print(f"recall score{recall_score(y_test,y_pred1,average='weighted')}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred2 = rf.predict(x_test)
```

```python
print(f"accuarcy{accuracy_score(y_test,y_pred2)}")
print(f"f1 score{f1_score(y_test,y_pred2,average='weighted')}")
print(f"recall score{recall_score(y_test,y_pred2,average='weighted')}")
```
